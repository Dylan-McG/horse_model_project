```python
# %% 
# üìä 07_visuals.ipynb ‚Äì Investor-Facing Charts and Dashboard
# ---------------------------------------------------------
# Loads summary data from 06 and visualises investor-grade outputs using src/visuals.py

# %%
# === Imports and Setup ===
import pandas as pd
from pathlib import Path
import sys
import yaml

# Add project root for package imports
PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(PROJECT_ROOT))

from src.visuals import (
    plot_field_size_histogram,
    plot_margin_distribution,
    plot_race_edge_map,
    plot_top_picks_barplot,
    plot_confidence_distribution,
    plot_mispriced_edge_barplot
)

# Paths
OUT_DIR = PROJECT_ROOT / "outputs"

# === CONFIG: Model Selection ===
CONFIG_PATH = PROJECT_ROOT / "config.yaml"
with open(CONFIG_PATH, "r") as f:
    config = yaml.safe_load(f)
print(f"üîç Visualizing outputs for model: {config['default_model']} ({config['prediction_files'][config['default_model']]})")

# (Optional) Suffix for outputs if you want per-model charts/CSVs:
# model_suffix = f"_{config['default_model']}" if per_model_outputs else ""

# %%
# === 1. Load Data ===
race_edge_path   = OUT_DIR / "summary_race_edge.csv"
top_safe_path    = OUT_DIR / "summary_top_safe.csv"
longshots_path   = OUT_DIR / "summary_longshots.csv"
conf_bins_path   = OUT_DIR / "summary_confidence_bins.csv"
mispriced_path   = OUT_DIR / "summary_top_mispriced.csv"

race_edge = pd.read_csv(race_edge_path)
top_safe = pd.read_csv(top_safe_path)
longshots = pd.read_csv(longshots_path)
conf_bins = pd.read_csv(conf_bins_path)
mispriced = pd.read_csv(mispriced_path)

# %%
# === 2. Field Size Distribution ===
plot_field_size_histogram(race_edge)
# Commentary: This chart shows the typical number of horses per race.

# %%
# === 3. Top-2 Margin Distribution ===
plot_margin_distribution(race_edge)
# Commentary: High margin = clear favourite; low margin = competitive race.

# %%
# === 4. Race Edge Map ===
plot_race_edge_map(race_edge)
# Commentary: Combines entropy and top-2 margin to highlight stable vs. chaotic races.

# %%
# === 5. Top 10 Safest Picks ===

# Patch for column name compatibility:
for df in [top_safe, longshots]:
    if "Pred_StackRidge" in df.columns:
        df.rename(columns={"Pred_StackRidge": "Predicted_Probability"}, inplace=True)
    elif "Pred_BlendAvg" in df.columns:
        df.rename(columns={"Pred_BlendAvg": "Predicted_Probability"}, inplace=True)

plot_top_picks_barplot(
    top_safe,
    title="Top 10 Highest Confidence Horses (Model Favourites)",
    dynamic_xlim=True
)
# Commentary: Horses with the highest predicted win probability.

# %%
# === 6. Top 10 Longshots (Hidden Gems) ===
plot_top_picks_barplot(
    longshots,
    title="Top 10 Attractive Longshots (Non-Favourites)",
    dynamic_xlim=True
)
# Commentary: Non-favourites with strong model-predicted probability.

# %%
# === 7. Confidence Distribution ===
plot_confidence_distribution(conf_bins)
# Commentary: Visualise the distribution of model confidence across all predictions.

# %%
# === 8. Mispriced Horses (Edge vs. Market) ===
plot_mispriced_edge_barplot(mispriced)
# Commentary: Horses where the model most disagrees with market pricing‚Äîpotential value bets.

# %%
print("\n‚úÖ Investor visual pack complete. Ready for PDF export or web dashboard integration.")



```
